```python
#1
from collections import Counter

def analyze_sentence(sentence):
    words = sentence.lower().split()

    result = {
        "word_count": len(words),
        "char_count": len(sentence.replace(" ", "")),
        "unique_words": len(set(words)),
        "most_frequent_word": Counter(words).most_common(1)[0][0]
    }
    return result

sentence = "The quick brown fox jumps over the lazy dog the fox"
print(analyze_sentence(sentence))
```

```python
#2
text = "A man a plan a canal Panama"

reverse_text = ""
for ch in text:
    reverse_text = ch + reverse_text
print("Reversed:", reverse_text)


cleaned = text.replace(" ", "").lower()
print("Palindrome:", cleaned == cleaned[::-1])


vowels = "aeiouAEIOU"
replaced = "".join("*" if ch in vowels else ch for ch in text)
print("Vowels Replaced:", replaced)


counts = {}
for v in "aeiou":
    counts[v] = text.lower().count(v)

print("Vowel Counts:", counts)
```

```python
#3
def format_report(data):
    print(f"{'Name':<15}{'Score':>6}{'Grade':^8}")
    print("-" * 29)

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [
    ("Alice", 92, "A"),
    ("Bob", 78, "B+"),
    ("Charlie", 55, "C"),
    ("Diana", 100, "A+")
]

format_report(data)
```

```python
#4
def list_ops(nums):
    unique = []
    for num in nums:
        if num not in unique:
            unique.append(num)

    even_squared = [nums[i]**2 for i in range(0, len(nums), 2)]

    nested = [[1,2,3],[4,5],[6,7,8,9]]
    flattened = [item for sublist in nested for item in sublist]

    return unique, even_squared, flattened

nums = [3,1,4,1,5,9,2,6,5,3,5]

unique, even_squared, flattened = list_ops(nums)

print("Unique:", unique)
print("Even Index Squared:", even_squared)
print("Flattened:", flattened)
```

```python
#7
def inventory_manager(records):
    result = {}

    for product, category, quantity, price in records:
        if category not in result:
            result[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        result[category]["total_items"] += quantity
        result[category]["total_value"] += quantity * price
        result[category]["products"].append(product)

    return result

records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20),
]

print(inventory_manager(records))
```

```python
#9 stats_utils.py
def mean(data):
    return sum(data) / len(data)

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)

def normalize(data):
    mn = min(data)
    mx = max(data)

    return [(x - mn) / (mx - mn) for x in data]
```

```python
#10 (closures)
def make_multiplier(n):
    def multiply(x):
        return x * n
    return multiply

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print(triple(10))
print(quintuple(10))
```

```python
#10 (Time decorator)
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()

        print("Execution Time:", end - start)
        return result

    return wrapper

@timer
def slow_sum(n):
    return sum(i*i for i in range(1, n+1))

print(slow_sum(100000))
```

```python
#10(Validation Decorator)
import math

def validate_positive(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if arg <= 0:
                raise ValueError("Arguments must be positive")
        return func(*args, **kwargs)

    return wrapper

@validate_positive
def safe_sqrt(x):
    return math.sqrt(x)

print(safe_sqrt(25))

# print(safe_sqrt(-4))  # Raises ValueError
```
