```python
#Q1.Vowel Counter
```

```python
def vowel_counter(text):
    vowels = {'a':0, 'e':0, 'i':0, 'o':0, 'u':0}
    total = 0

    for ch in text.lower():
        if ch in vowels:
            vowels[ch] += 1
            total += 1

    print("Total Vowels:", total)
    print("Vowel Counts:", vowels)

vowel_counter("Hello Python Programming")
```

```python
#Q2.Palindrome Checker with List
```

```python
def palindrome_check(text):
    reverse_list = []

    for i in range(len(text)-1, -1, -1):
        reverse_list.append(text[i])

    reverse_text = "".join(reverse_list)

    if text == reverse_text:
        print("Palindrome")
    else:
        print("Not a Palindrome")

palindrome_check("madam")
```

```python
#Q3.Character Frequency Counter
```

```python
def char_frequency(text):
    freq = {}

    for ch in text:
        if ch in freq:
            freq[ch] += 1
        else:
            freq[ch] = 1

    print(freq)

char_frequency("banana")
```

```python
#Q4.Word Length Classification
```

```python
def classify_words(sentence):
    words = sentence.split()

    short_words = []
    medium_words = []
    long_words = []

    for word in words:
        if len(word) < 5:
            short_words.append(word)
        elif len(word) <= 8:
            medium_words.append(word)
        else:
            long_words.append(word)

    print("Short Words:", short_words)
    print("Medium Words:", medium_words)
    print("Long Words:", long_words)

classify_words("Python is powerful and programming language")
```

```python
#Q5.Substrings with Specific Start/End
```

```python
def matching_strings(words):
    result = []

    for word in words:
        if word.startswith("py") and word.endswith("thon"):
            result.append(word)

    print("Matching Strings:", result)

data = ["python", "pythons", "pyathon", "pyxxthon"]
matching_strings(data)
```

```python
#Q6.Unique Word Extractor
```

```python
import string

def unique_words(paragraph):
    paragraph = paragraph.lower()

    for p in string.punctuation:
        paragraph = paragraph.replace(p, "")

    unique = set()

    for word in paragraph.split():
        unique.add(word)

    result = sorted(list(unique))
    print(result)

unique_words("Python is easy. Python is powerful!")
```

```python
#Q7.Even/Odd Indexed Character Separator
```

```python
def separate_characters(text):
    even_chars = []
    odd_chars = []

    for index, char in enumerate(text):
        if index % 2 == 0:
            even_chars.append(char)
        else:
            odd_chars.append(char)

    print("Even Index Characters:", even_chars)
    print("Odd Index Characters:", odd_chars)

separate_characters("Python")
```

```python
#Q8.String Anagram Checker
```

```python
def anagram_check(str1, str2):
    list1 = list(str1)
    list2 = list(str2)

    list1.sort()
    list2.sort()

    if list1 == list2:
        print("Anagrams")
    else:
        print("Not Anagrams")

anagram_check("listen", "silent")
```

```python
#Q9.Missing Character Finder (Alphabet Check)
```

```python
def missing_letters(text):
    alphabets = set("abcdefghijklmnopqrstuvwxyz")

    for ch in text.lower():
        if ch in alphabets:
            alphabets.remove(ch)

    print("Missing Letters:", alphabets)

missing_letters("hello world")
```

```python
#Q10.Custom String Reversal (Word by Word)
```

```python
def reverse_words(sentence):
    words = sentence.split()

    reversed_words = []

    for i in range(len(words)-1, -1, -1):
        reversed_words.append(words[i])

    result = " ".join(reversed_words)
    print(result)

reverse_words("Hello World")
```
