```python
st = input("enter a string: ")
ch1 = 'a'
ch2 = 'g'
for ch in [ch1, ch2]:
  found = False

  for i in st:
      if i == ch:
        found = True
        break
  if found:
    print(ch, "is present in the string")
  else:
    print(ch, "is not present in the string")
```

```python
text = input("enter teh text:")
count = 0
vowel_dict = {}
for i in text.lower():
  if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
    count += 1
    if i in vowel_dict:
      vowel_dict[i] += 1
    else:
      vowel_dict[i] = 1

print("total vowels: ",count)
print("vowels counts: ",vowel_dict)
```

```python
def palindrome_checker(text):
    rev_list = []

    for ch in text:
        rev_list.insert(0, ch)

    rev_text = "".join(rev_list)

    if text == rev_text:
        print("Palindrome")
    else:
        print("Not a palindrome")

# Example
palindrome_checker("madam")
```
